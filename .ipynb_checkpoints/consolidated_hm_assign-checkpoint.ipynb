````python
def handle_statlog_german_credit_data(statlog_german_credit_data):
    """
    Transform Statlog German Credit dataset into a pandas DataFrame.

    Parameters:
    - statlog_german_credit_data: object
        The Statlog German Credit dataset object containing features, targets, and variable descriptions.

    Returns:
    - data: pandas DataFrame
        The dataset as a pandas DataFrame with appropriate column names based on variable descriptions.
    """
    # Extract features and targets
    X = statlog_german_credit_data.data.features
    y = statlog_german_credit_data.data.targets
    
    # Extract variable names and descriptions
    vars = statlog_german_credit_data.variables
    cols_map = dict(vars[['name', 'description']].values)
    
    # Combine features and targets into a DataFrame
    data = pd.concat([X, y], axis=1)
    
    # Rename columns based on variable descriptions
    data.rename(columns=cols_map, inplace=True)
    
    return data

def determine_variable_type(df, ratio_threshold=0.05):
    """
    Determine whether each column in the DataFrame contains discrete or continuous variables.

    Parameters:
    - df: pandas DataFrame
        The dataset to analyze.
    - ratio_threshold: float, optional (default=0.05)
        The threshold ratio to determine whether a variable is discrete or continuous.

    Returns:
    - variable_types: dict
        A dictionary where keys are column names and values are the variable types (Discrete or Continuous).
    """
    variable_types = {}
    for column in df.columns:
        unique_values = df[column].nunique()
        total_values = len(df[column])
        ratio = unique_values / total_values
        if ratio < ratio_threshold:
            variable_type = 'Discrete'
            summary = describe_discrete_variable(df[column])
            # Plot the value counts if the variable is discrete
            fig = plot_categorical_variable_counts(df, column, title=f'{column} Distribution', ylabel='Frequency')
            plt.show()
            # Print summary statistics for discrete variables
            print(f"Column: {column}")
            print(f"Variable type: {variable_type}")
            print("Summary:")
            print(f"    Unique values: {summary['unique_values']}")
            print("    Value counts:")
            for value, count in summary['value_counts'].items():
                print(f"        {value}: {count}")
            print("\n----------------------\n")
        else:
            variable_type = 'Continuous'
            summary = describe_continuous_variable(df[column])
            # Print summary statistics for continuous variables
            print(f"Column: {column}")
            print(f"Variable type: {variable_type}")
            print("Summary:")
            print(f"    Mean: {summary['mean']}")
            print(f"    Median: {summary['median']}")
            print(f"    Standard deviation: {summary['std_dev']}")
            print(f"    Minimum: {summary['min']}")
            print(f"    Maximum: {summary['max']}")
            print("\n----------------------\n")
        
        variable_types[column] = variable_type
        
    return variable_types


def describe_discrete_variable(series):
    """
    Describe a discrete variable.

    Parameters:
    - series: pandas Series
        The column to analyze.

    Returns:
    - summary: dict
        A dictionary containing the summary of the discrete variable.
    """
    unique_values = series.nunique()
    summary = {
        'type': 'Discrete',
        'unique_values': unique_values,
        'value_counts': series.value_counts().to_dict()
    }
    return summary

def describe_continuous_variable(series):
    """
    Describe a continuous variable.

    Parameters:
    - series: pandas Series
        The column to analyze.

    Returns:
    - summary: dict
        A dictionary containing the summary of the continuous variable.
    """
    summary = {
        'type': 'Continuous',
        'mean': series.mean(),
        'median': series.median(),
        'std_dev': series.std(),
        'min': series.min(),
        'max': series.max()
    }
    return summary


def plot_categorical_variable_counts(data, column_name, title=None, ylabel=None):
    """
    Plot the percentage counts of a categorical variable as a bar plot.

    Parameters:
    - data: pandas DataFrame
        The dataset containing the variable to plot.
    - column_name: str
        The name of the column containing the categorical variable.
    - title: str, optional
        The title of the plot.
    - ylabel: str, optional
        The label for the y-axis.

    Returns:
    - fig: matplotlib Figure
        The plotted figure.
    """
    total_count = len(data[column_name])
    counts = data[column_name].value_counts(normalize=True) * 100  # Calculate percentage counts
    fig = counts.plot.bar()
    fig.axhline(y=5, color='red')
    fig.axhline(y=10, color='orange')
    if title:
        fig.set_title(title)
    if ylabel:
        fig.set_ylabel(ylabel)
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()
    # Adjust layout to prevent overlapping labels
    for i, count in enumerate(counts):
        plt.text(i, count + 0.5, f'{count:.1f}%', ha='center', va='bottom')  # Add percentage labels
    return fig

def filter_categorical_columns(var_types, target, ignore_cat=[]):
    """
    Filter categorical columns based on specified conditions.

    Parameters:
    - var_types: dict
        A dictionary where keys are column names and values are variable types (Discrete or Continuous).
    - target: str
        The name of the target variable.
    - ignore_cat: list, optional
        A list of column names to ignore.

    Returns:
    - cat_cols: list
        A list of categorical columns based on the specified conditions.
    """
    cat_cols = [col for col, v_type in var_types.items() 
                if v_type == 'Discrete' and col != target and col not in ignore_cat]
    return cat_cols

def filter_continuous_columns(var_types, ignore_cat=[]):
    """
    Filter continuous columns based on specified conditions.

    Parameters:
    - var_types: dict
        A dictionary where keys are column names and values are variable types (Discrete or Continuous).
    - ignore_cat: list, optional
        A list of column names to ignore.

    Returns:
    - cont_cols: list
        A list of continuous columns based on the specified conditions.
    """
    cont_cols = [col for col, v_type in var_types.items() if v_type == 'Continuous']
    cont_cols.extend(ignore_cat)
    return cont_cols
    return cat_cols

def calculate_mean_target_per_category(df, var, target):
    """
    Calculate the mean target value per category of a variable.

    Parameters:
    - df: pandas DataFrame
        The dataset containing the variables.
    - var: str
        The name of the variable for which to calculate the mean target value per category.
    - target: str
        The name of the target variable.

    Returns:
    - temp_df: pandas DataFrame
        A DataFrame containing the percentage of entries per category and the mean target values.
    """
    total_entries = len(df)

    # Calculate the percentage of entries per category
    perc_entries = df[var].value_counts() / total_entries
    temp_df = pd.DataFrame({var: perc_entries.index, 'perc_entries': perc_entries.values})

    # Add the mean target values
    mean_target = df.groupby(var)[target].mean().reset_index()
    pos_count = df.groupby([var, target]).count().reset_index()
    temp_df = temp_df.merge(mean_target, on=var, how='left')

    return temp_df

def plot_categories(df, var, target):
    """
    Plot the percentage of entries per category and the average target value per category.

    Parameters:
    - df: pandas DataFrame
        The dataset containing the variables.
    - var: str
        The name of the variable for which to plot the categories.
    - target: str
        The name of the target variable.

    Returns:
    - None
        Displays the plot.
    """
    fig, ax = plt.subplots(figsize=(8, 4))
    plt.xticks(df.index, df[var], rotation=90)

    ax2 = ax.twinx()
    ax.bar(df.index, df["perc_entries"], color='lightgrey')
    ax2.plot(df.index, df[target], color='green', label='Seconds')
    ax.axhline(y=0.05, color='red')
    ax.set_ylabel('Percentage of entries per category')
    ax.set_xlabel(var)
    ax2.set_ylabel('Average target value per category')
    plt.show()

def group_rare_labels(df, var, threshold=0.05):
    """
    Group rare labels in a categorical variable.

    Parameters:
    - df: pandas DataFrame
        The dataset containing the variable.
    - var: str
        The name of the variable for which to group rare labels.
    - threshold: float, optional (default=0.05)
        The threshold below which categories are considered rare.

    Returns:
    - tmp: pandas Series
        A Series with rare labels replaced by combined categories.
    """
    total_entries = len(df)
    # Calculate the percentage of counts per category
    category_percentages = df[var].value_counts() / total_entries

    # Identify rare labels
    rare_labels = category_percentages[category_percentages < threshold].index.tolist()

    # Replace rare labels with combined category string
    combined_category = '_'.join(rare_labels)
    tmp = df[var].replace(rare_labels, combined_category)
    return tmp

def group_categories(df, var, categories_to_group, new_category_name):
    """
    Group one or more categories within a feature variable.

    Parameters:
    - df: pandas DataFrame
        The dataset containing the variable.
    - var: str
        The name of the variable for which to group categories.
    - categories_to_group: list
        A list of categories to group together.
    - new_category_name: str
        The name of the new category after grouping.

    Returns:
    - tmp: pandas Series
        A Series with specified categories grouped into a new category.
    """
    # Replace specified categories with new category name
    tmp = df[var].replace(categories_to_group, new_category_name)
    return tmp


def woe_discrete(df, discrete_variable_name, good_bad_variable_df):
    """
    Calculate Weight of Evidence (WoE) for a discrete variable.

    Parameters:
    - df: pandas DataFrame
        The dataset containing the discrete variable and the good/bad indicator variable.
    - discrete_variable_name: str
        The name of the discrete variable.
    - good_bad_variable_df: pandas DataFrame
        A DataFrame containing the good/bad indicator variable.

    Returns:
    - df: pandas DataFrame
        A DataFrame containing WoE values and related statistics for the discrete variable.
    """
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis=1)
    df = pd.concat([
        df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].count(),
        df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].mean()], axis=1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop=True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df

def plot_by_woe(df_WoE, rotation_of_x_axis_labels=0):
    """
    Plot Weight of Evidence (WoE) by category.

    Parameters:
    - df_WoE: pandas DataFrame
        A DataFrame containing WoE values for categories.
    - rotation_of_x_axis_labels: int, optional (default=0)
        Rotation angle for x-axis labels.

    Returns:
    - None
        Displays the plot.
    """
    x = np.array(df_WoE.iloc[:, 0].apply(str))
    y = df_WoE['WoE']
    plt.figure(figsize=(18, 6))
    plt.plot(x, y, marker='o', linestyle='--', color='k')
    plt.xlabel(df_WoE.columns[0])
    plt.ylabel('Weight of Evidence')
    plt.title('Weight of Evidence by ' + df_WoE.columns[0])
    plt.xticks(rotation=rotation_of_x_axis_labels)
    plt.show()
    
def woe_ordered_continuous(df, discrete_variabe_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variabe_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df

def chisq_of_df_cols(df, c1, c2):
    """
    Calculates the chi-square statistic between two categorical variables in a DataFrame.

    Parameters:
    - df (DataFrame): The DataFrame containing the data.
    - c1 (str): The name of the first categorical variable.
    - c2 (str): The name of the second categorical variable.

    Returns:
    - chi2 (float): The chi-square statistic.
    - p (float): The p-value.
    - dof (int): The degrees of freedom.
    - expected (ndarray): The expected frequencies, based on the marginal sums of the contingency table.
    """
    groupsizes = df.groupby([c1, c2]).size()
    ctsum = groupsizes.unstack(c1)
    # fillna(0) is necessary to remove any NAs which will cause exceptions
    return chi2_contingency(ctsum.fillna(0))

# Function to plot histogram and print chi-square statistic for ordinal variable
def plot_and_chi2(df, ordinal_var, target_var):
    # Define the order of labels for the ordinal variable
    order = ordinals_w_orders[ordinal_var]
    
    # Reorder the ordinal variable in the DataFrame
    temp_df = df.copy()
    temp_df[ordinal_var] = pd.Categorical(temp_df[ordinal_var], order)
    
    # Plot histogram
    sns.histplot(
        binwidth=0.5,
        x=ordinal_var,
        hue=target_var,
        data=df,
        stat="density",
        multiple="dodge",
    )
    
    plt.xlabel(ordinal_var)
    plt.ylabel('Probability')
    plt.title(f'Histogram of {ordinal_var} by {target_var}')
    plt.show()
    
    # Calculate chi-square statistic
    chi2, p, dof, expected = chisq_of_df_cols(df, c1=target_var, c2=ordinal_var)
    
    # Print chi-square statistic
    print(f"Chi-square test between {ordinal_var} and {target_var}:")
    print(f"Chi-square statistic: {chi2}")
    print(f"p-value: {p}")
    print(f"Degrees of freedom: {dof}")
    print("Expected frequencies:")
    print(expected)

def plot_and_chi2_nominal(df, nominal_var, target_var):
    # Plot histogram
    sns.histplot(
        binwidth=0.5,
        x=nominal_var,
        hue=target_var,
        data=df,
        stat="density",
        multiple="dodge",
    )
    
    plt.xlabel(nominal_var)
    plt.ylabel('Probability')
    plt.title(f'Histogram of {nominal_var} by {target_var}')
    plt.show()
    
    # Calculate chi-square statistic
    chi2, p, dof, expected = chisq_of_df_cols(df, c1=target_var, c2=nominal_var)
    
    # Print chi-square statistic
    print(f"Chi-square test between {nominal_var} and {target_var}:")
    print(f"Chi-square statistic: {chi2}")
    print(f"p-value: {p}")
    print(f"Degrees of freedom: {dof}")
    print("Expected frequencies:")
    print(expected)

def plot_continuous_vs_categorical(df, continuous_var, categorical_var):
    # Calculate point-biserial correlation coefficient and p-value
    pb_corr, p_value = pointbiserialr(df[continuous_var], df[categorical_var])
    
    # Plot boxplot to visualize the distribution of the continuous variable by categories of the target variable
    sns.boxplot(x=categorical_var, y=continuous_var, data=df)
    plt.xlabel(categorical_var)
    plt.ylabel(continuous_var)
    plt.title(f'Boxplot of {continuous_var} by {categorical_var}\nPoint-Biserial Correlation: {pb_corr}, p-value: {p_value}')
    plt.show()

    # Print correlation coefficient and p-value
    # print(f"Point-Biserial Correlation between '{continuous_var}' and '{categorical_var}': {pb_corr:.2f}")
    # print(f"P-value: {p_value:.4f}")

def define_pipeline(classifier):
    """Define the modeling pipeline."""
    pipeline = Pipeline([
        # ('scaler', StandardScaler()),  # Standardize features if needed
        ('classifier', classifier)  # Classifier
    ])
    return pipeline

def train_model(pipeline, X_train, y_train):
    """Train the model."""
    pipeline.fit(X_train, y_train)

def validate_model(pipeline, X_test, y_test):
    """Validate the model and calculate AUC-ROC score."""
    y_prob = pipeline.predict_proba(X_test)[:, 1]
    auc_roc_score = roc_auc_score(y_test, y_prob)
    
    return auc_roc_score

def plot_roc_curve(fpr, tpr, auc_roc_score):
    """Plot ROC curve."""
    plt.plot(fpr, tpr, label='ROC Curve (area = %0.2f)' % auc_roc_score)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()

def evaluate_model(y_test, y_pred):
    """Calculate confusion matrix, precision, recall, F1-score, accuracy, TP, TN, FP, FN."""
    conf_matrix = confusion_matrix(y_test, y_pred)
    TN, FP, FN, TP = conf_matrix.ravel()
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    return conf_matrix, precision, recall, f1, accuracy, TP, TN, FP, FN
````

# Discretizing Continuous Variables for Weight of Evidence (WoE) Analysis

## Insight
Continuous variables have been discretized into bins to facilitate Weight of Evidence (WoE) analysis. The following variables have been discretized along with their specified number of bins:
- 'Credit amount' into 10 bins
- 'Age' into 5 bins
- 'Duration' into 10 bins

Discretization allows for the transformation of continuous variables into categorical variables, enabling the application of WoE analysis. By discretizing continuous variables, patterns and relationships with the target variable can be identified and visualized using WoE plots, aiding in the interpretation of predictive modeling results.


````python
# Dictionary defining continuous variables and their corresponding number of bins for discretization
cut_cont_cols = {
    'Credit amount': 10,
    'Age': 5,
    'Duration': 10,
}

# Looping through each continuous variable and its corresponding number of bins for discretization
for cont_col in cut_cont_cols:
    # Discretize the current continuous variable into bins
    df_inputs_prepr[f'{cont_col}_factor'] = pd.cut(df_inputs_prepr[cont_col], cut_cont_cols[cont_col])
    
    # Calculate Weight of Evidence (WoE) for the discretized variable
    df_temp = woe_ordered_continuous(df_inputs_prepr, f'{cont_col}_factor', df_targets_prepr)
    
    # Plot Weight of Evidence (WoE) by category for the discretized variable
    plot_by_woe(df_temp)
````
